## Agent ์ํ

๊ทธ๋ํ๋ฅผ ์์ํ๊ฒ์ต๋๋ค.

๊ฐ ๋ธ๋์ ์๋ฌ๋๋ `state` ๊ฐ์ฒด์๋๋ค.

์ํ๋ `messages` ๋ชฉ๋ก์ผ๋ก ๊ตฌ์ฑ๋ฉ๋๋ค.

๊ทธ๋ํ์ ๊ฐ ๋ธ๋๋ ์ด ๋ชฉ๋ก์ ๋ด์ฉ์ ์ถ๊ฐํฉ๋๋ค.

```python
from typing import Annotated, Sequence, TypedDict
from langchain_core.messages import BaseMessage
from langgraph.graph.message import add_messages


# ์์ด์ํธ ์ํ๋ฅผ ์์ํ๋ ํ์ ๋์๋๋ฆฌ, ๋ฉ์์ง ์ํ์ค๋ฅผ ๊ด๋ฆฌํ๊ณ ์ถ๊ฐ ๋์ ์์
class AgentState(TypedDict):
    # add_messages reducer ํจ์๋ฅผ ์ฌ์ฉํ์ฌ ๋ฉ์์ง ์ํ์ค๋ฅผ ๊ด๋ฆฌ
    messages: Annotated[Sequence[BaseMessage], add_messages]
```

## ๋ธ๋์ ์ฃ์ง

์์ด์ํธ ๊ธฐ๋ฐ RAG ๊ทธ๋ํ๋ ๋ค์๊ณผ ๊ฐ์ด ๊ตฌ์ฑ๋ ์ ์์ต๋๋ค.

* **์ํ**๋ ๋ฉ์์ง๋ค์ ์งํฉ์๋๋ค
* ๊ฐ **๋ธ๋**๋ ์ํ๋ฅผ ์๋ฐ์ดํธ(์ถ๊ฐ)ํฉ๋๋ค
* **์กฐ๊ฑด๋ถ ์ฃ์ง**๋ ๋ค์์ ๋ฐฉ๋ฌธํ ๋ธ๋๋ฅผ ๊ฒฐ์ํฉ๋๋ค

๊ฐ๋จํ ์ฑ์๊ธฐ(Grader)๋ฅผ ๋ง๋ค์ด ๋ณด๊ฒ์ต๋๋ค.

```python
from typing import Literal
from langchain import hub
from langchain_core.messages import HumanMessage
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from pydantic import BaseModel, Field
from langchain_openai import ChatOpenAI
from langgraph.prebuilt import tools_condition
from langchain_teddynote.models import get_model_name, LLMs

# ์ต์ ๋ชจ๋ธ์ด๋ฆ ๊ฐ์ธ์ค๊ธฐ
MODEL_NAME = get_model_name(LLMs.GPT4)


# ๋ฐ์ดํฐ ๋ชจ๋ธ ์์
class grade(BaseModel):
    """A binary score for relevance checks"""

    binary_score: str = Field(
        description="Response 'yes' if the document is relevant to the question or 'no' if it is not."
    )


def grade_documents(state) -> Literal["generate", "rewrite"]:
    # LLM ๋ชจ๋ธ ์ด๊ธฐํ
    model = ChatOpenAI(temperature=0, model=MODEL_NAME, streaming=True)

    # ๊ตฌ์กฐํ๋ ์ถ๋ฅ์ ์ํ LLM ์ค์
    llm_with_tool = model.with_structured_output(grade)

    # ํ๋กฌํํธ ํํ๋ฆฟ ์์
    prompt = PromptTemplate(
        template="""You are a grader assessing relevance of a retrieved document to a user question. \n
        Here is the retrieved document: \n\n {context} \n\n
        Here is the user question: {question} \n
        If the document contains keyword(s) or semantic meaning related to the user question, grade it as relevant. \n
        Give a binary score 'yes' or 'no' score to indicate whether the document is relevant to the question.""",
        input_variables=["context", "question"],
    )

    # llm + tool ๋ฐ์ธ๋ฉ ์ฒด์ธ ์์ฑ
    chain = prompt | llm_with_tool

    # ํ์ฌ ์ํ์์ ๋ฉ์์ง ์ถ์ถ
    messages = state["messages"]

    # ๊ฐ์ฅ ๋ง์ง๋ง ๋ฉ์์ง ์ถ์ถ
    last_message = messages[-1]

    # ์๋ ์ง๋ฌธ ์ถ์ถ
    question = messages[0].content

    # ๊ฒ์๋ ๋ฌธ์ ์ถ์ถ
    retrieved_docs = last_message.content

    # ๊ด๋จ์ฑ ํ๊ฐ ์คํ
    scored_result = chain.invoke({"question": question, "context": retrieved_docs})

    # ๊ด๋จ์ฑ ์ฌ๋ถ ์ถ์ถ
    score = scored_result.binary_score

    # ๊ด๋จ์ฑ ์ฌ๋ถ์ ๋ฐ๋ฅธ ๊ฒฐ์
    if score == "yes":
        print("==== [DECISION: DOCS RELEVANT] ====")
        return "generate"

    else:
        print("==== [DECISION: DOCS NOT RELEVANT] ====")
        print(score)
        return "rewrite"


def agent(state):
    # ํ์ฌ ์ํ์์ ๋ฉ์์ง ์ถ์ถ
    messages = state["messages"]

    # LLM ๋ชจ๋ธ ์ด๊ธฐํ
    model = ChatOpenAI(temperature=0, streaming=True, model=MODEL_NAME)

    # retriever tool ๋ฐ์ธ๋ฉ
    model = model.bind_tools(tools)

    # ์์ด์ํธ ์๋ต ์์ฑ
    response = model.invoke(messages)

    # ๊ธฐ์กด ๋ฆฌ์คํธ์ ์ถ๊ฐ๋๋ฏ๋ก ๋ฆฌ์คํธ ํํ๋ก ๋ฐํ
    return {"messages": [response]}


def rewrite(state):
    print("==== [QUERY REWRITE] ====")
    # ํ์ฌ ์ํ์์ ๋ฉ์์ง ์ถ์ถ
    messages = state["messages"]
    # ์๋ ์ง๋ฌธ ์ถ์ถ
    question = messages[0].content

    # ์ง๋ฌธ ๊ฐ์์ ์ํ ํ๋กฌํํธ ๊ตฌ์ฑ
    msg = [
        HumanMessage(
            content=f""" \n
    Look at the input and try to reason about the underlying semantic intent / meaning. \n
    Here is the initial question:
    \n ------- \n
    {question}
    \n ------- \n
    Formulate an improved question: """,
        )
    ]

    # LLM ๋ชจ๋ธ๋ก ์ง๋ฌธ ๊ฐ์
    model = ChatOpenAI(temperature=0, model=MODEL_NAME, streaming=True)
    # Query-Transform ์ฒด์ธ ์คํ
    response = model.invoke(msg)

    # ์ฌ์์ฑ๋ ์ง๋ฌธ ๋ฐํ
    return {"messages": [response]}


def generate(state):
    # ํ์ฌ ์ํ์์ ๋ฉ์์ง ์ถ์ถ
    messages = state["messages"]

    # ์๋ ์ง๋ฌธ ์ถ์ถ
    question = messages[0].content

    # ๊ฐ์ฅ ๋ง์ง๋ง ๋ฉ์์ง ์ถ์ถ
    docs = messages[-1].content

    # RAG ํ๋กฌํํธ ํํ๋ฆฟ ๊ฐ์ธ์ค๊ธฐ
    prompt = hub.pull("teddynote/rag-prompt")

    # LLM ๋ชจ๋ธ ์ด๊ธฐํ
    llm = ChatOpenAI(model_name=MODEL_NAME, temperature=0, streaming=True)

    # RAG ์ฒด์ธ ๊ตฌ์ฑ
    rag_chain = prompt | llm | StrOutputParser()

    # ๋ต๋ณ ์์ฑ ์คํ
    response = rag_chain.invoke({"context": docs, "question": question})
    return {"messages": [response]}
```

## ๊ทธ๋ํ

* `call_model` ์์ด์ํธ๋ก ์์ํฉ๋๋ค
* ์์ด์ํธ๊ฐ ํจ์๋ฅผ ํธ์ถํ์ง ๊ฒฐ์ํฉ๋๋ค
* ํจ์ ํธ์ถ์ ๊ฒฐ์ํ ๊ฒฝ์ฐ, ๋๊ตฌ(retriever)๋ฅผ ํธ์ถํ๊ธฐ ์ํ `action`์ ์คํํฉ๋๋ค
* ๋๊ตฌ์ ์ถ๋ฅ๊ฐ์ ๋ฉ์์ง(`state`)์ ์ถ๊ฐํ์ฌ ์์ด์ํธ๋ฅผ ํธ์ถํฉ๋๋ค

```python
from langgraph.graph import END, StateGraph, START
from langgraph.prebuilt import ToolNode
from langgraph.checkpoint.memory import MemorySaver

# AgentState ๊ธฐ๋ฐ ์ํ ๊ทธ๋ํ ์ํฌํ๋ก์ฐ ์ด๊ธฐํ
workflow = StateGraph(AgentState)

# ๋ธ๋ ์์
workflow.add_node("agent", agent)  # ์์ด์ํธ ๋ธ๋
retrieve = ToolNode([retriever_tool])
workflow.add_node("retrieve", retrieve)  # ๊ฒ์ ๋ธ๋
workflow.add_node("rewrite", rewrite)  # ์ง๋ฌธ ์ฌ์์ฑ ๋ธ๋
workflow.add_node("generate", generate)  # ๊ด๋จ ๋ฌธ์ ํ์ธ ํ ์๋ต ์์ฑ ๋ธ๋

# ์ฃ์ง ์ฐ๊ฒฐ
workflow.add_edge(START, "agent")

# ๊ฒ์ ์ฌ๋ถ ๊ฒฐ์์ ์ํ ์กฐ๊ฑด๋ถ ์ฃ์ง ์ถ๊ฐ
workflow.add_conditional_edges(
    "agent",
    # ์์ด์ํธ ๊ฒฐ์ ํ๊ฐ
    tools_condition,
    {
        # ์กฐ๊ฑด ์ถ๋ฅ์ ๊ทธ๋ํ ๋ธ๋์ ๋งคํ
        "tools": "retrieve",
        END: END,
    },
)

# ์ก์ ๋ธ๋ ์คํ ํ ์ฒ๋ฆฌ๋ ์ฃ์ง ์์
workflow.add_conditional_edges(
    "retrieve",
    # ๋ฌธ์ ํ์ง ํ๊ฐ
    grade_documents,
)
workflow.add_edge("generate", END)
workflow.add_edge("rewrite", "agent")

# ๊ทธ๋ํ ์ปดํ์ผ
graph = workflow.compile(checkpointer=MemorySaver())
```
